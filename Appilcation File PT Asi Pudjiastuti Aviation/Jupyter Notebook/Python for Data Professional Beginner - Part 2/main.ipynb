## List Manipulation

```python
# Fitur .append()
print(">>> .append()")
list_makanan = ["Gado-gado","Ayam Goreng","Rendang"]
list_makanan.append("Ketoprak")
print(list_makanan)

# Fitur .clear()
print("\n>>> .clear()")
list_makanan = ["Gado-gado","Ayam Goreng","Rendang"]
list_makanan.clear()
print(list_makanan)

# Fitur .copy()
print("\n>>> .copy()")
list_makanan1 = ["Gado-gado","Ayam Goreng","Rendang"]
list_makanan2 = list_makanan1.copy()
list_makanan3 = list_makanan1
list_makanan2.append("Opor")
list_makanan3.append("Ketoprak")
print(list_makanan1)
print(list_makanan2)
print(list_makanan3)

# Fitur .count()
print("\n>>> .count()")
list_score = ['Budi', 'Sud', 'Budi', 'Budi', 'Budi', 'Sud', 'Sud']
score_budi = list_score.count("Budi")
score_sud = list_score.count("Sud")
print(score_budi)
print(score_sud)

# Fitur .extend()
print("\n>>> .extend()")
list_menu = ['Gado-gado', 'Ayam Goreng', 'Rendang']
list_minuman = ['Es Teh', 'Es Jeruk', 'Es Campur']
list_menu.extend(list_minuman)
print(list_menu)

# Fitur .index()
print("\n>>> .index()")
list_score = ['Budi','Sud','Budi','Budi','Budi','Sud','Sud']
score_pertama_sud = list_score.index("Sud")
print(score_pertama_sud)

# Fitur .insert()
print("\n>>> Fitur .insert()")
list_score = ['Budi','Sud','Budi','Budi','Budi','Sud','Sud']
list_score.insert(3, "XXX")
print(list_score)

# Fitur .pop()
print("\n>>> Fitur .pop()")
list_menu = ['Gado-gado', 'Ayam Goreng', 'Rendang', 'Ketoprak']
list_menu.pop(0)
list_menu.pop(2)
print(list_menu)

# Fitur .remove()
print("\n>>> Fitur .remove()")
list_menu = ['Gado-gado', 'Ayam Goreng', 'Rendang', 'Ketoprak']
list_menu.remove("Gado-gado")
list_menu.remove("Ketoprak")
print(list_menu)

# Fitur .reverse()
print("\n>>> .reverse()")
list_menu = ['Gado-gado', 'Ayam Goreng', 'Rendang', 'Ketoprak']
list_menu.reverse()
print(list_menu)

# Fitur .sort()
print("\n>> .sort()")
list_menu = ['Gado-gado', 'Ayam Goreng', 'Rendang', 'Ketoprak']
list_menu.sort()
print(list_menu)
list_menu.sort(reverse=True)
print(list_menu)
```

## Tuple Manipulation

```python
# Fitur .count()
print(">>> Fitur .count()")
tuple_score = ("Budi", "Sud", "Budi", "Budi", "Budi", "Sud", "Sud")
score_budi = tuple_score.count("Budi")
score_sud = tuple_score.count("Sud")
print(score_budi)
print(score_sud)
# Fitur .index()
tuple_score = ("Budi", "Sud", "Budi", "Budi", "Budi", "Sud", "Sud")
score_pertama_sud = tuple_score.index("Sud") + 1
print(score_pertama_sud)
```
