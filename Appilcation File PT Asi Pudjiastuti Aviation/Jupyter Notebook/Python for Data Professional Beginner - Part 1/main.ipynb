## Sequence Type -> List [Mutable]

```python
contoh_list = [1, "dua", 3, 4.0, 5]
print(contoh_list[0])
print(contoh_list[3])
contoh_list = [1, "dua", 3, 4.0, 5]
contoh_list[3] = "empat"
print(contoh_list)
```

## Sequence Type -> Tuple (Imutable)

```python
contoh_tuple = ("Januari", "Februari", "Maret", "April")
print(contoh_tuple[0])
contoh_tuple = ("Januari", "Februari", "Maret", "April")
contoh_tuple[0] ="Desember" # Error karena Imutable
```

## Set Type -> Data harus Unique & Tidak memperhatikan urutan

```python
contoh_list = ["Dewi", "Budi", "Cici", "Linda", "Cici"]
print(contoh_list)
contoh_set = {"Dewi", "Budi", "Cici", "Linda", "Cici"}
print(contoh_set)
contoh_frozen_set = ({"Dewi", "Budi", "Cici", "Linda", "Cici"})
print(contoh_frozen_set)
```

```python
sepatu = {"nama" : "Sepatu Niko",
          "harga" : 150000,
          "diskon" : 3000}
baju = {"nama" : "Baju Unikloh",
        "harga" : 80000,
        "diskon" : 8000}
celana = {"nama" : "Celana Lepis",
         "harga" : 200000,
         "diskon" : 60000}
daftar_belanja = [sepatu, baju, celana]
print(daftar_belanja)
```

```python
# 1. Data yang dinyatakan dalam bentuk dictionary
sepatu = {"nama" : "Sepatu Niko",
          "harga" : 150000,
          "diskon" : 3000}
baju = {"nama" : "Baju Unikloh",
        "harga" : 80000,
        "diskon" : 8000}
celana = {"nama" : "Celana Lepis",
         "harga" : 200000,
         "diskon" : 60000}
# 2. Hitung harga masing-masing data setelah dikurangi diskon
harga_sepatu = sepatu["harga"] - sepatu["diskon"]
harga_baju = baju["harga"] - baju["diskon"]
harga_celana = celana["harga"] - celana["diskon"]
# 3. Hitung harga total 
harga_total = harga_sepatu + harga_baju + harga_celana
# 4. Hitung harga kena pajak
total_pajak = harga_total * 0.1
# 5. Cetak harga_total + total_pajak
print(harga_total + total_pajak)
```
