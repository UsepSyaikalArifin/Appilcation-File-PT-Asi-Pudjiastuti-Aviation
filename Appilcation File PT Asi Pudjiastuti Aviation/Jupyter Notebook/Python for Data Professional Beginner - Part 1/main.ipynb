```python
# Tagihan
tagihan = [50000, 75000, 12500, 300000, 200000]
# Tanpa menggunakan while loop
total_tagihan = tagihan[0] + tagihan[1] + tagihan[2] + tagihan[3] + tagihan[4]
print(total_tagihan)

# Menggunakan while loop
i = 0
jumlah_tagihan = len(tagihan)
total_tagihan = 0
while i < jumlah_tagihan:
    total_tagihan += tagihan[i]
    i += 1
print(total_tagihan)
```

```python
tagihan = [50000, 75000, -150000, 125000, 300000, -50000, 200000]
i = 0
jumlah_tagihan = len(tagihan)
total_tagihan = 0
while i < jumlah_tagihan:
    if tagihan[i] < 0:
        total_tagihan = -1
        print("Terdapat angka minus dalam tagihan, perhitungan dihentikan!")
        break
    total_tagihan += tagihan[i]
    i += 1
print(total_tagihan)
```

```python
tagihan = [50000, 75000, -150000, 125000, 300000, -50000, 200000]
i = 0
jumlah_tagihan = len(tagihan)
total_tagihan = 0
while i < jumlah_tagihan:
    if tagihan[i] < 0:
        i += 1
        continue
    total_tagihan += tagihan[i]
    i += 1
print(total_tagihan)
```

# Tugas

```python
uang_jalan = 1500000
jumlah_hari = 31
list_plat_nomor = [8993, 2198, 2501, 2735, 3772, 4837, 9152]

kendaraan_genap = 0
kendaraan_ganjil = 0
for kendaraan in list_plat_nomor:
    if kendaraan % 2 == 0:
        kendaraan_genap += 1
    else:
        kendaraan_ganjil += 1
print(kendaraan_genap)
print(kendaraan_ganjil)

hari = 0
total_pengeluaran_ganjil = 0
total_pengeluaran_genap = 0
while hari < jumlah_hari:
    if hari % 2 == 0:
        total_pengeluaran_genap += (kendaraan_genap * uang_jalan)
    else:
        total_pengeluaran_ganjil += (kendaraan_ganjil * uang_jalan)
    hari += 1

total_pengeluaran = total_pengeluaran_genap + total_pengeluaran_ganjil
print(total_pengeluaran_genap)
print(total_pengeluaran_ganjil)
print(total_pengeluaran)
```
