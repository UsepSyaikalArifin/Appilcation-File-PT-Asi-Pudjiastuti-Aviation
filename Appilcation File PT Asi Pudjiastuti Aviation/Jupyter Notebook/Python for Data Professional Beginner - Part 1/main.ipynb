## Sequence Type -> List [Mutable]

```python
contoh_list = [1, "dua", 3, 4.0, 5]
print(contoh_list[0])
print(contoh_list[3])
contoh_list = [1, "dua", 3, 4.0, 5]
contoh_list[3] = "empat"
print(contoh_list)
```

## Sequence Type -> Tuple (Imutable)

```python
contoh_tuple = ("Januari", "Februari", "Maret", "April")
print(contoh_tuple[0])
contoh_tuple = ("Januari", "Februari", "Maret", "April")
contoh_tuple[0] ="Desember" # Error karena Imutable
```
